```python
allocated = list()
with open("allocs2022B.csv",'r') as f:
    for line in f:
        student = line.strip().split(',')[0]
        allocated.append(student)
 
master_list = list()
with open("student_list",'r') as f:
    for line in f:
        student = line.strip()
        master_list.append(student)
        
for s in master_list:
    if not s in allocated:
        print(s+"@columbia.edu",end=',')
```
